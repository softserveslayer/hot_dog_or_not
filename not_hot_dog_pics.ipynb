```python
import urllib.request
import os
import pandas as pd
```

```python
df = pd.read_csv('http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07696728', skiprows = [396,671] , header=None)
for idx, row in df.iterrows():
    print(row[0])
    try:
        filename = os.path.join('data/not_hot_dog', f'{idx}-a.jpg')
        urllib.request.urlretrieve(row[0], filename)
    except:
        pass
```
